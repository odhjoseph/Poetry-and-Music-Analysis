```python
import json
%matplotlib inline
```

```python
with open("gutenberg-metadata.json") as json_data:
    gutenbergData = json.load(json_data)
    json_data.close()
```

```python
genre = []
for IDs in gutenbergData:
    genre.append(IDs.get('LoC Class'))
```

```python
#double checked for ones without colons
genrelist = []
shortnedGenre = []
for item in genre:
    if(type(item) is not None):
        for subitem in item:
            splice = subitem.find(":")
            genrelist.append(subitem[splice + 1:])
            if(":" in subitem[splice + 2:]):
                shortnedGenre.append(subitem[splice + 2:])
                
```

```python
collectionsList = []
index = 0
for genre in shortnedGenre:
    if ":" not in genre:
        collectionsList.append(genre)
    else:
        index = genre.find(":")
        collectionsList.append(genre[:index])
```

```python
import collections
import matplotlib.pyplot as plt
```

```python
counts = collections.Counter(collectionsList)
```

```python
labels = []
sizes = []
for item in counts:
    labels.append(item)
    sizes.append(counts[item])
    
patches, texts = plt.pie(sizes, shadow=False, startangle=150)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.show()
```

```python
def contains_digit(s):
    isdigit = str.isdigit
    return any(map(isdigit,s))
```

```python
yearsPoetry = []
for item in shortnedGenre:
    if contains_digit(item):
        findparenth = item.find("(")
        yearsPoetry.append(item[findparenth + 1:len(item) -1])
```

```python
yearsCounter = collections.Counter(yearsPoetry)
yearsCounter
```

```python
labels = []
sizes = []
for item in yearsCounter:
    labels.append(item)
    sizes.append(yearsCounter[item])
    
patches, texts = plt.pie(sizes, shadow=False, startangle=150)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.show()
```
